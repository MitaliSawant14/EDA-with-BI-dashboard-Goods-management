```python
#bad expiration

df[df["expiration_date"] < df["transaction_date"]]
```

```python
# bad pricing

df[(df["selling_price"] < 0) | (df["cost_price"] < 0) | (df["base_price"] < 0)]
```

```python
# bad quantity

df[(df["units_sold"] > df["initial_quantity"]) | (df["units_wasted"] > df["initial_quantity"])]
```

```python
df["profit_margin"] = np.where(
    df["revenue"] != 0,
    df["profit"] / df["revenue"],
    0
)
```

```python
df["waste_rate"] = (df["units_wasted"] - df["initial_quantity"])
```

```python
df["days_to_expiry"] = (df["expiration_date"] - df["transaction_date"]).dt.days
```

```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
# Product Storage temperature

px.histogram(
    df,
    x='storage_temp',
    nbins=120,
    color_discrete_sequence=['#EF553B'],
    title='Storage Temperature Distribution',
).show()
```

```python
# product count by category

fig = px.bar(
    df['category'].value_counts().reset_index(),
    x='count',
    y='category',
    orientation='h',
    color='category',
    title='Product Count by Category'
)

fig.update_xaxes(
    tickvals=[3000, 6000, 9000,],
    ticktext=['3k', '6k', '9k'],
    range=[0, 15000],
    title='Count'
)

fig.show()
```

```python
df.head()
```

```python
# Monthly Trends

monthly = df.groupby("month", as_index=False)[
    ["revenue", "profit", "waste_cost"]
].sum()

fig = px.line(
    monthly,
    x="month",
    y=["revenue", "profit", "waste_cost"],
    title="Monthly Business Trends",
    markers=True
)

fig.update_layout(
    xaxis_title="Month",
    yaxis_title="Amount",
    legend_title="Metric"
)

fig.show()
```

```python
# Monthly revenue

monthly_revenue = df.groupby("month", as_index=False)["revenue"].sum()

fig = px.line(
    monthly_revenue,
    x="month",
    y="revenue",
    title="Revenue by Month",
    markers=True
)

fig.update_layout(xaxis_title="Month", yaxis_title="Revenue")
fig.show()
```

```python
# monthly profit

monthly_profit = df.groupby("month", as_index=False)["profit"].sum()

fig = px.line(
    monthly_profit,
    x="month",
    y="profit",
    title="Profit by Month",
    markers=True
)

fig.update_layout(xaxis_title="Month", yaxis_title="Profit")
fig.show()
```

```python
# monthly waste

monthly_waste = df.groupby("month", as_index=False)["waste_cost"].sum()

fig = px.line(
    monthly_waste,
    x="month",

    y="waste_cost",
    title="Waste Cost by Month",
    markers=True
)

fig.update_layout(xaxis_title="Month", yaxis_title="Waste Cost")
fig.show()
```

```python
# profit by category

cat_profit = df.groupby("category", as_index=False)["profit"].sum().sort_values("profit", ascending=False)
fig = px.bar(cat_profit, x="profit", y="category", color = 'category', orientation="h", title="Total Profit by Category")
fig.show()
```

```python
# waste by category

waste_by_cat = df.groupby("category")["waste_cost"].sum().sort_values(ascending=False)
px.bar(waste_by_cat, title="Waste Cost by Category").show()
```

```python
# lifespan by category

lifespan_by_cat = df.groupby("category")["shelf_life_days"].sum().sort_values(ascending=False)
px.bar(lifespan_by_cat, title="shelf_life_days by Category").show()
```
